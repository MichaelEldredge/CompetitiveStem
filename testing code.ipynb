```python
# Do not modify this code
from random import shuffle

mylist = [1,7,3,4,5,6,2]

# swaps and comparisons are just counters we will use to measure how efficient your program is
swaps = 0
comparisons = 0

# This function will swap the a and b entries of a list
def swap(a,b):
    global swaps
    swaps += 1
    temp = mylist[a]
    mylist[a] = mylist[b]
    mylist[b] = temp

# This function will return true if the a entry is less than or equal to the b entry, false otherwise
def leq(a,b):
    global comparisons
    comparisons += 1
    return mylist[a] <= mylist[b]
```

```python
mylist = list(range(100))
shuffle(mylist)
swaps = 0
comparisons = 0



# copy and paste your sort here
for i in range(len(mylist)):
    for j in range(i,len(mylist)):
        if leq(j,i):
            swap(i,j)



print(swaps)
print(comparisons)
if mylist == list(range(len(mylist))):
    print("Good job. The list is sorted")
else:
    print("This is not sorted")
    print(mylist)
```
