# While loops, For loops, and the Range function

Computers are really good at doing similar tasked over and over again. In this section we will explore how we can tell a computer to do that using loops.

The while loop will run a block of code until a condition returns false.

```python
# This prints out random numbers up to 100
from random import randint
from time import sleep
# randint(a,b) will return a random integer between 1 and 20.
# sleep(1) will pause the program for one second

x = randint(1,20)
while x <= 100:
    print(x)
    sleep(1)
    x = x + randint(1,20)
```

A common problem that programmers need to deal with is infinite loops. Consider what would have happened if I had forgotten the line "x = x + randint(1,20)" in the above program.

```python
x = randint(1,20)
while x <= 100:
    print(x)
    sleep(1)
    # This is an infinite loop. You will need to stop it yourself.
```

A for loop is another type of loop. Instead of using a condition, it does something to all the elements of a list.

```python
fib = [1,1,2,3,5,8]

for number in fib:
    print(number)
    sleep(1)
```

The range() function is a way of generating a list of numbers.

```python
for number in range(5):
    print(number)
```

In the cell below I use a loop to demonstrate a way you might use loops to sort a list

```python
mylist = [ 14, 67,   38,   86,   91,   69,   93,   64,    8,   16]

swaps = 0
comparisons = 0

# This function will swap the a and b entries of a list
def swap(a,b):
    global swaps
    swaps += 1
    temp = mylist[a]
    mylist[a] = mylist[b]
    mylist[b] = temp

# This function will return true if the a entry is less than or equal to the b entry, false otherwise
def compare(a,b):
    global comparisons
    comparisons += 1
    return mylist[a] <= mylist[b]

for i in range(10):
    if compare(i,0):
        swap(i,0)

print(mylist)
```

Notice how the lowest number in the list (that is 8) has been moved to the front of the list.
