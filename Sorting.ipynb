Your assignment will be to sort a list that we provide, using the swap and comparison operators that are provided below.

```python
# Do not modify this code

mylist = [1,7,3,4,5,6,2]

# swaps and comparisons are just counters we will use to measure how efficient your program is
swaps = 0
comparisons = 0

# This function will swap the a and b entries of a list
def swap(a,b):
    global swaps
    swaps += 1
    temp = mylist[a]
    mylist[a] = mylist[b]
    mylist[b] = temp

# This function will return true if the a entry is less than or equal to the b entry, false otherwise
def compare(a,b):
    global comparisons
    comparisons += 1
    return mylist[a] <= mylist[b]


print("The old list is as follows.")
print(mylist)

swap(1,6)

print("The new list after the swap is:")
print(mylist)
```

```python
print(compare(1,2))
print(compare(2,1))
```

# Conditionals
 
You can tell the computer to do different things depending on a condition using if statements

```python
if True:
    print("This will be printed")

if 1 > 2:
    print("This is not printed")
    
```

```python
if False:
    print("You can also use else statements to do something if you have two paths for the code.")
    print("These lines are not going to be printed.")
else:
    print("You can see this line was printed")
```
